```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df2[~df2['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqt_num(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
    
```

```python
df2
```

```python
df3=df2.copy()
df3['total_sqft']=df3['total_sqft'].apply(convert_sqt_num)
```

```python
df4=df3.copy()
df4['price_per_sqft']=df4['price']*100000/df4['total_sqft']
df4.head()
```

```python
df4.location=df4.location.apply(lambda x:x.strip())#to remove white spaces front n back
location_stats=df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
df5=df4[~(df4.total_sqft/df4.bhk<300)]
df5.shape
```

```python
 df5.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df=subdf[(subdf.price_per_sqft>(m-st))&(subdf.price_per_sqft<(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df6=remove_pps_outliers(df5)
df6
```

```python
df6.shape
```

```python
def plot_scatter_chat(df,location):
    bhk2=df[(df.location==location) & (df.bhk==2)]
    bhk3=df[(df.location==location) & (df.bhk==3)]
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',marker='+')
    plt.scatter(bhk3.total_sqft,bhk3.price,color='red',marker='*')
    plt.xlabel('total_sqft')
    plt.ylabel('price')
    plt.legend()
plot_scatter_chat(df6,'Hebbal')
    
```

```python
def remove_bhk_outliers(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk,bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df7=remove_bhk_outliers(df6)
df7.shape
```

```python
df8=df7[df7.bath<df7.bhk+2]
df8.shape
```

```python
df9=df8.drop(['price_per_sqft','size'],axis='columns')
df9
```

```python
dummies=pd.get_dummies(df9.location)
dummies=dummies.astype('int')
dummies
```

```python
df10=pd.concat([df9,dummies.drop('other',axis='columns')],axis='columns')
df10.head()
```

```python
df11=df10.drop('location',axis='columns')
df11
```

```python
X=df11.drop('price',axis='columns')
y=df11.price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=10)
cross_val_score(LinearRegression(),X,y,cv=cv)
```

```python
def predict_price(location,sqft,bath,bhk):
    loc_index=np.where(X.columns==location)[0][0]
    x=np.zeros(len(X.columns))
    x[0]=sqft
    x[1]=bath
    x[2]=bhk 
    if loc_index>=0:
        x[loc_index]=1
    return lr.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000,2,2)
```

```python
predict_price('Whitefield',20000,5,8)
```
